```python
import pandas as pd
import re
```

```python
f = open("WhatsApp Chat with AI.txt", "r", encoding = "utf-8")
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r"\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}[\s\u202f]*[ap]m\s-\s"
```

```python
msgs = re.split(pattern, data)[1:]
msgs
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({"user_msgs" : msgs, "date" : dates})
df["date"] = pd.to_datetime(df["date"], format = "%d/%m/%y, %I:%Mâ€¯%p - ")

df
```

```python
df.shape
```

```python
#separate users and msgs
users = []
msgss = []
for msgs in df["user_msgs"]:
  i = re.split(r"([^:]+):\s(.*)", msgs)
  if i[1:]:
    users.append(i[1])
    msgss.append(i[2])
  else:
    users.append("group_notification")
    msgss.append(i[0])

df["users"] = users
df["msgs"] = msgss
df.drop(["user_msgs"], axis = 1, inplace = True)
df
```

```python
df["year"] = df["date"].dt.year
df.head()
```

```python
df["month"] = df["date"].dt.month_name()
df.head()
```

```python
df["day"] = df["date"].dt.day
df.head()
```

```python
df["hour"] = df["date"].dt.hour
df["minute"] = df["date"].dt.minute
df.head()
```

```python
#total no. of msgs --> df.shape
```

```python
#total no. of words
words = []
for i in df["msgs"]:
  words.extend(i.split(' '))

len(words)
```

```python
#total no. of media shared
num_media = df[df["msgs"] == "<Media omitted>"]
len(num_media)
```

```python
#top users
top_users = df[df["users"] != "group_notification"]["users"].value_counts().head()
```

```python
import matplotlib.pyplot as plt
```

```python
name = top_users.index
count = top_users.values
plt.bar(name, count)
plt.xticks(rotation = "vertical")
plt.show()
```

```python
import emoji
emojis = []
for i in df["msgs"]:
  emojis.extend([c for c in i if emoji.is_emoji(c)])

emojis
```

```python
from collections import Counter
pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis)))).rename(columns={0: 'Emoji', 1: 'Count'})
```

```python
df["month_num"] = df["date"].dt.month
df
```

```python
timeline = df.groupby(["year", "month_num", "month"]).count()["msgs"].reset_index()
timeline
```

```python
time = []
for i in range(timeline.shape[0]):
  time.append(timeline["month"][i] + "-" + str(timeline["year"][i]))
```

```python
timeline["time"] = time
timeline
```

```python
plt.plot(timeline["time"], timeline["msgs"])
plt.xticks(rotation = "vertical")
plt.show()
```
